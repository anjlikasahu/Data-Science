```python
knowing_stats=cg['UndergradMajor'].apply(lambda x: x.str.contains('statistics').sum())
knowing_stats
```

```python
stats_df=pd.concat([country_respondents , knowing_stats ], axis ='columns')

stats_df
```

```python
#renaming the columns in the above table:
stats_df.rename(columns={'Country' : 'total_people' , 'UndergradMajor' : 'knowing_stats'} , inplace=True)
```

```python
#adding another column named percentage:
stats_df['percentage'] = (stats_df['knowing_stats']/stats_df['total_people']) *100 
```

```python
stats_df.sort_values(by='percentage' , ascending=False,inplace=True)
stats_df
```
